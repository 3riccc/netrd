### Quick example of clustering/reordering a adjacency matrix

```python
import matplotlib.pyplot as plt
from netrd.reconstruction import TimeGrangerCausalityReconstructor
from netrd.utilities import cluster, read
import networkx as nx
```

```python
TS = read.read_time_series("../data/synth_4clique_N64_simple.csv")
```

```python
recon = TimeGrangerCausalityReconstructor()
G = recon.fit(TS)
```

### Plot the unordered matrix

```python
adj = nx.to_numpy_matrix(G)
plt.imshow(adj)
plt.colorbar()
plt.show()
```

### Cluster the graph
#### Returns ordered matrix, the new index order, dendogram dictionary, and linkage

```python
adjClustered, order, dend, link = cluster.clusterGraph(G)
```

```python
plt.imshow(adjClustered)
plt.colorbar()
plt.show()
```
